```python
# Define a function to categorize iPhone models based on their characteristics
def categorize_iphone_model(model):
    model = model.lower()
    if "pro max" in model:
        return "pro max"
    elif "pro" in model:
        return "pro"
    elif "plus" in model:
        return "plus"
    else:
        return "standard"

# Apply the function to the 'modelo' column and create a new 'category' column
df['category'] = df['modelo'].apply(categorize_iphone_model)

# Display the updated dataset with the new 'category' column
df.head()
```
