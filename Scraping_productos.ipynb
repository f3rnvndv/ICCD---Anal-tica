```python
import requests
from bs4 import BeautifulSoup
import csv

def get_product_details(url, model_selector, brand_selector, normal_price_selector, offer_price_selector):
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Error al acceder a la p√°gina: {response.status_code}")
        return []

    soup = BeautifulSoup(response.text, 'html.parser')
    products = soup.select(model_selector)

    product_details = []
    for product in products:

        model = product.text.strip()

        brand_element = product.find_next('b', class_=brand_selector)
        brand = brand_element.text.strip() if brand_element else "No disponible"

        normal_price_element = product.find_next('span', class_=normal_price_selector)
        normal_price = normal_price_element.text.strip() if normal_price_element else "No disponible"

        offer_price_element = product.find_next('span', class_=offer_price_selector)
        offer_price = offer_price_element.text.strip() if offer_price_element else "No disponible"

        product_details.append({
            'Modelo': model,
            'Precio Normal': normal_price,
            'Precio Oferta': offer_price
        })

    return product_details

base_url = "https://XXXXXXXXX" #Aca poner link de la pagina a realizar scraping
start_indices = range(0, 80, 40)  

all_product_details = []
for start in start_indices:
    url = base_url.format(start)
    print(f"Scraping {url}")
    product_details = get_product_details(
        url,
        # Extaer la estructura del HTML dependiendo de la informacion a sacar 
        model_selector='XXXXXXXXXXXXXXXXXXXXX', 
        brand_selector='XXXXXXXXXXXXXXXXXXXXX',
        normal_price_selector='XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX',
        offer_price_selector='XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'
    )
    all_product_details.extend(product_details)

csv_file = "productos.csv"

with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["Modelo", "Precio Normal", "Precio Oferta"])
    writer.writeheader()  
    for product in all_product_details:
        writer.writerow(product)  

print(f"Data guardada en {csv_file}")
```
